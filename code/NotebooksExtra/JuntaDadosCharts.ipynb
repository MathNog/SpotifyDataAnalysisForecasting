União de Diferentes planilhas do Spotify Charts

```python
import pandas as pd
import numpy as np
```

```python
# path padrão para todos os arquivos csv do Spotify Charts
path = "/content/drive/MyDrive/INF1032 - Spotify/"

#Funcao para carregar e concatenas os diferentes arquivos do spotify charts
def carrega_dados_charts(path):   

  #Nomes dos arquivos TOP 200 diário global 
  lista_paths = ["regional-global-daily-2022-03-27.csv",
                 "regional-global-daily-2022-04-01.csv",
                 "regional-global-daily-2022-04-02.csv"]

  #DataFrame vazio para armazenar a concatenacao de dataframes
  df_charts = pd.DataFrame()

  #Vamos concatenar cada csv que possuímos
  for arq in lista_paths:
    df = pd.read_csv(path+arq)

    #Renomeando as colunas com a primeira linha e dropando essa primeira linha
    df.columns = df.iloc[0,:]
    df = df.drop(axis=0,index=0)
    df = df.reset_index(drop=True)

    #Pegando apenas a data do nome de cada arquivo e adicionando uma nova coluna com essa informacao
    #CUIDADO -> quando fiz download do arquivo mais atual ele veio com o nome "latest" alguma coisa. Eu, manualmente, renomeei o arquivo csv para ter a data
    #Precisamos ter os nomes padronizados para essa linha abaixo fazer sentido e sempre pegar uma data como slice do nome do arquivo
    data = arq[-14:-4]
    df['data'] = data
    df['data'] = pd.to_datetime(df['data'])

    #Concatenando os datasets um embaixo do outro
    df_charts = pd.concat([df_charts,df])
    #O shape[0] deve sempre aumentar de 200 em 200
    print(df_charts.shape)
  
  #Resetar os indices
  df_charts = df_charts.reset_index(drop=True)
  return df_charts

```

```python
df_charts = carrega_dados_charts(path)
```

```python
df_charts.to_csv(path + "charts-top200-daily-global.csv")
```
