```python
# pegar todos os csv encontrados em uma pasta, passando o caminho dela
def get_all_csv (path):
  all_files = glob.glob(os.path.join(path + "/*.csv"))

  li = []
  for filename in all_files:
      # skiprows=1 pois na primeira linha possui o comentário 
      # "Note that these figures are generated using a formula that protects against any 
      # artificial inflation of chart positions." 
      df = pd.read_csv(filename, skiprows=1)

      # pegar o país de acordo com o nome do arquivo csv
      keyword = 'regional-'
      before_keyword, keyword, after_keyword = filename.partition(keyword)
      region_code = after_keyword[:2]
      # se o código não estiver na lista de países disponíveis é porque
      # o top 200 é global
      if region_code.upper() in countries_available:
        df["Region"] = countries_available.get(region_code.upper())

        keyword_2 = "regional-" + region_code + "-weekly-"
        before_keyword, keyword_2, after_keyword = filename.partition(keyword_2)
        week = os.path.splitext(after_keyword)[0]
        
        if week == "latest":
          df["Start_date"] = datetime.datetime.strptime("2022-05-06", "%Y-%m-%d").date()
          # df["End_date"] = datetime.datetime.strptime("2022-05-13", "%Y-%m-%d").date()
        else:
          keyword_3 = "--"
          before_keyword, keyword_3, after_keyword = week.partition(keyword_3)
          df["Start_date"] = datetime.datetime.strptime(before_keyword, "%Y-%m-%d").date()
          # df["End_date"] = datetime.datetime.strptime(after_keyword, "%Y-%m-%d").date()
      else:
        df["Region"] = "Global"

        keyword_2 = "regional-global-weekly-"
        before_keyword, keyword_2, after_keyword = filename.partition(keyword_2)
        week = os.path.splitext(after_keyword)[0]

        if week == "latest":
          df["Start_date"] = datetime.datetime.strptime("2022-05-06", "%Y-%m-%d").date()
          # df["End_date"] = datetime.datetime.strptime("2022-05-13", "%Y-%m-%d").date()
        else:
          keyword_3 = "--"
          before_keyword, keyword_3, after_keyword = week.partition(keyword_3)
          df["Start_date"] = datetime.datetime.strptime(before_keyword, "%Y-%m-%d").date()
          # df["End_date"] = datetime.datetime.strptime(after_keyword, "%Y-%m-%d").date()

      track_ids_list =[]
      for index, row in df.iterrows():
        track_ids_list.append(row['URL'].split('/')[-1].split('?')[0])
      df["Track_id"] = track_ids_list

      li.append(df)

  df = pd.concat(li, axis=0, ignore_index=True)

  return df
```
